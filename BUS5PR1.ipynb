```python
import pandas as pd #importing panda for reading data.
```

```python
df = pd.read_csv("/content/2-wheeler-EV-bikewale.csv")#taking data from dataline path.
df.head()
```

```python
df.dropna(subset=['review'], inplace=True)#using dropna to delete thye null value.
print(df)
```

```python
df.columns #printing the coloum
```

```python
#function of removing numbers
import string
def removing_number(text):
  trans = str.maketrans('','','1234567890')
  return text.translate(trans)
removing_number("a1b2c3d")
```

```python
df['review'] = df['review'].apply(removing_number)#applying the previous number.
df['review'][0]#Print the review coloum 1
```

```python
#function of removing punctuation
import string
def removing_punctuation(text):
  trans = str.maketrans('','',string.punctuation)
  return text.translate(trans)
removing_punctuation("a,b,c,d")
```

```python
df['review'] = df['review'].apply(removing_punctuation)#Applying the previous removing punctuation function
df['review'][0]
```

```python
#removing stop word
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords#using stop word inbuild function.
swords=stopwords.words('english')
```

```python
def remove_stopwords(text):#Using the stop word function.
    return" ".join([word.lower() for word in str(text).split() if word not in swords])
df['review']=df['review'].apply(remove_stopwords)
df['review'][0]
```

```python
#Analysis word frequency
from nltk.tokenize import  word_tokenize
from nltk.stem import WordNetLemmatizer
from nltk.stem import PorterStemmer
nltk.download('punkt')
nltk.download('wordnet')

```

```python
Word_lemmatizer = WordNetLemmatizer()
all_text=" ".join(df['review'])
word_token= word_tokenize(all_text)#use inbuild functions for tokenization.
print(len(word_token))
lemma=WordNetLemmatizer()#Using the inbuild importing function to lematize data.
word_token=[lemma.lemmatize(word) for word in word_token]
ps=PorterStemmer()
word_token=[ps.stem(word) for word in word_token]
```

```python
from collections import Counter
word_int=Counter(word_token)
word_data = pd.DataFrame(word_int.items(),columns=["word",'count'])#Dividing colums into 2 parameters word and count
word_data = word_data.sort_values(by='count',ascending=False)
word_data
```

```python
example = df['review'][50]
print(example)#print 5oth review line
```

```python
tokens = nltk.word_tokenize(example)
tokens[:20]
```

```python
nltk.download('averaged_perceptron_tagger')
tagged = nltk.pos_tag(tokens)
tagged[:10]
```

```python
nltk.download('words')
nltk.download('maxent_ne_chunker')
entities = nltk.chunk.ne_chunk(tagged)
entities.pprint()
```

```python
#Sentiment analysis on Roberta Pretrained Model
!pip install transformers
from transformers import pipeline #Importing pileline
```

```python
MODEL = f"cardiffnlp/twitter-roberta-base-sentiment"
task_of_sentiment = pipeline("sentiment-analysis", model=MODEL, tokenizer=MODEL)#importing pipeline funtion
```

```python
df['review'][0]#Print 1st review coloum data
```

```python
task_of_sentiment(df['review'][0])#Examine first data and giving result.
```

```python
df['review'][1]#Print 2nd review coloum data
```

```python
task_of_sentiment(df['review'][1])#Examine first data and giving result.
```

```python
def sentiment_analysis(text):#Creating a loop to purse whole review coloum and examine
  try:
    sen=task_of_sentiment(text)[0]['label']
  except:
    sen=None

  return sen
df['sentiment']=df['review'].apply(sentiment_analysis)
df
```

Count plot on sentiment analysis

```python
# count plot on sentiment analysis
sns.countplot(y ='sentiment',data = df)
# Show the plot on the trained data
plt.show()
```
